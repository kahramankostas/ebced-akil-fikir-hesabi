# hesaplamayı yapan fonksiyonlar

```python
def print_af(akil,fikir,a,b,burcu):
    
    burc={1:"KOÇ-Merih",
    2:"BOĞA-Zühre",
    3:"İKİZLER-Utarid",
    4:"YENGEÇ-Ay",
    5:"ASLAN-Güneş",
    6:"BAŞAK-Utarid",
    7:"TERAZİ-Zühre",
    8:"AKREP-Merih",
    9:"YAY-Müşteri",
    10:"OĞLAK-Zühal",
    11:"KOVA-Zühal",
    0:"BALIK-Müşteri"}

    a=a%9
    b=b%9
    if abs(a-b)!=0:
        uyum="Eşlerin İsim Uyumu : "+str(abs(a-b))
    else:
        uyum="Eşlerin İsim Uyumu : 9"
    burcu= "Burç ve Yıldızı : "+ str(burc[burcu])    
    lbl = tk.Label(root, text=uyum,font=("Arial Bold", 9))
    lbl.grid(column=0, row=8)
    lbl = tk.Label(root, text=akil,font=("Arial Bold", 9))
    lbl.grid(column=0, row=9)
    lbl = tk.Label(root, text=fikir,font=("Arial Bold", 9))
    lbl.grid(column=0, row=10)
    
    lbl = tk.Label(root, text=burcu,font=("Arial Bold", 9))
    lbl.grid(column=0, row=11)

def correlations(Anne, 
                 Baba,
                Çocuk_isim1,
                Çocuk_isim2,
                Çocuk_isim3,):
    listem=list(df["isim"])
    ebced_1=list(df["ebced"])
    if int(ebced_1[listem.index(Çocuk_isim1)])!=0 or int(ebced_1[listem.index(Çocuk_isim2)])!=0 or int(ebced_1[listem.index(Çocuk_isim3)])!=0 :
        akil=int(ebced_1[listem.index(Anne)])+int(ebced_1[listem.index(Çocuk_isim1)])+int(ebced_1[listem.index(Çocuk_isim2)])+int(ebced_1[listem.index(Çocuk_isim3)])
        fikir=int(ebced_1[listem.index(Baba)])+int(ebced_1[listem.index(Çocuk_isim1)])+int(ebced_1[listem.index(Çocuk_isim2)])+int(ebced_1[listem.index(Çocuk_isim3)])
        burcu=akil%12
        akil=akil%9
        fikir=fikir%9
        if akil==0:
            akil=9
        if fikir==0:
            fikir=9   
        akil= "Çocuğun Akıl Puanı : " + str(akil)
        fikir= "Çocuğun Fikir Puanı : " + str(fikir)
        print_af(akil,fikir,int(ebced_1[listem.index(Anne)]),int(ebced_1[listem.index(Baba)]),burcu)
    else:
        akil= "Çocuğun Akıl Puanı : 0"
        fikir= "Çocuğun Fikir Puanı : 0"
        print_af(akil,fikir,int(ebced_1[listem.index(Anne)]),int(ebced_1[listem.index(Baba)]),burcu)

    
        
        
def on_select(event=None):
    dic={}
    if event: # <-- this works only with bind because `command=` doesn't send event
        print("event.widget:", event.widget.get())
    for i, x in enumerate(all_comboboxes):
        #print("all_comboboxes[%d]: %s" % (i, x.get()))
        dic[i]=x.get()
    correlations(dic[0],dic[1],dic[2],dic[3],dic[4])
```

# Python Tk arayüzü

```python
akil= ""
fikir= ""
buton_isimleri={0:"Anne İsmi : ",1:"Baba İsmi : ",2:"Çocuk İsmi 1:  ",3:"Çocuk İsmi 2:  ", 4:"Çocuk İsmi 3 :  "}
names=[ "Anne İsmi","Baba İsmi","Çocuk İsmi 1","Çocuk İsmi 2", "Çocuk İsmi 3"]
root = tk.Tk()
root.geometry('550x300')
all_comboboxes = []
#tk.Label(root, text="")


root.title('Ebced ile çocuğun akıl fikir hesabı') 

     
# label text for title 
ttk.Label(root, text = "Anne, Baba ve Çocuk ismi seçiniz",  
  foreground ="black",  
          font = ("Times New Roman", 15)).grid(row = 0, column = 1) 
  
lbl = tk.Label(root, text="Eşlerin İsim Uyumu : 0" ,font=("Arial Bold", 9))
lbl.grid(column=0, row=8)
lbl = tk.Label(root, text="Çocuğun Akıl Puanı : 0",font=("Arial Bold", 9))
lbl.grid(column=0, row=9)
lbl = tk.Label(root, text="Çocuğun Fikir Puanı : 0 ",font=("Arial Bold", 9))
lbl.grid(column=0, row=10)
lbl = tk.Label(root, text="Çocuğun burç ve yıldızı : ",font=("Arial Bold", 9))
lbl.grid(column=0, row=11)

for i,ii in enumerate(names) :
    lbl = tk.Label(root, text=ii)
    lbl.grid(column=0, row=i+1)
    cb = ttk.Combobox(root, values=multiple)
    cb.set("seçiniz")
    cb.grid(column=1, row=i+1)
    all_comboboxes.append(cb)    

b = tk.Button(root, text="Akıl Fikir Hesapla", command=on_select)
b.grid(column=2, row=15)
root.mainloop()
```
